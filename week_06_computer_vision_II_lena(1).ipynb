#### Chapter 3.1 - Data Augmentation

**What is data augmentation?**
Have you ever lost your glasses and then squinted, or tried to look through a rainy window? Or looked at a false color image, maybe a forest where the trees are blue and the sky green? You can usually make an educated guess what you are looking at, even though the image you see is different than usual. This is, essentialy, what data augmentation is. It's the same data, still recognizable, but slightly altered in some way.

**What is that useful for?**
Let me begin with an anecdote that you've probably heard in the lectures. Say you have pictures of cats and dogs, and want your model to tell the two apart. How many people you know go to the park with their dogs? I imagine many. Hence, many images of dogs are dogs lying on the grass. The same is generally untrue for cats, at least I have never heard of anyone walking their cat to the park. At any rate, here is what happens when I train a neural network on these images: The model takes a shortcut. It sees a lot of green and the correct answer for these pictures is always "Dog". It learns "Green = Dog". This is what we call overfitting. We have overfitted to the existence of green in the background as a quintessential part of what makes a dog. Sometimes, we get away with this, if our data always has this correlation.

Now I get some new data. A bunch of people have taken pictures of their cats, sunbathing on the terrace. The garden is in the background. Lots of green. The model, in its infinite wisdom, will at first guess that these images are of dogs. Clearly, our model's ability to tell apart cats and dogs has not generalized to this new dataset.

So how can we prevent the model from taking shortcuts and encourage learning information that generalizes? We force these generalizations in training. If I gave you an image of a dog, but the grass was brown, and the dog green, you could still identify it as a dog, instead of a cat, right? And so should the model, if we can manage it. So let's also make it train using pictures of cats and dogs where the colours are different or removed. Suddenly, the shortcut solution is no longer useful, and the model must rely on shape, texture, or contextual information like a leash. The practice of color change described is a practical and useful data augmentation that is used in state-of-the-art image recognition.

In addition to color changes, there is a myriad of other techniques, such as cropping, image rotation or flipping, edge filters, solarization, random noise, and many, many more. Basically, anything that you believe may eventually show up in testing data and that you want the model to generalize to, can be made into a corresponding data augmentation.

**How do we use data augmentations in practice?**
There are two ways of adding data augmentation during training. Either, you can implement it inside of your dataset, so that it only returns augmented image tensors, or right before feeding your image tensors into your model. Both options are acceptable and come with advantages and disadvantages, although the more common way is to separate dataset and augmentations. We also showcase the native PyTorch way of augmenting data below.

If you are particularly eager, or want to try your hand at making image augmentation functions yourself, it can be fun and is definitely good practice. However, PyTorch comes with a large selection of image augmentations right out of the box, and in the following chapter, we will look at how to make use of them.

```python
import torch
import torch.nn as nn
import torchvision.transforms as tt
import torchvision.transforms.functional as ttf

# Torchvision contains two ways of utilizing transforms:
# Functional and Composed.

# Functional does what it advertises - it is a function which
# you can use on your tensors. Here is an example which performs
# a center crop:

dummy_images = torch.rand((16, 1, 256, 256))

transformed_images = ttf.center_crop(img = dummy_images, output_size = (128, 128))
print(transformed_images.size())

# Functional transforms have the inherent advantage of giving the
# user very fine-grained control.
```

```python
# The alternative is the so-called Composed form, which uses
# classes to achieve the same result. We make a Composed Transform
# like so:

dummy_images = torch.rand((16, 1, 256, 256))
transforms = tt.Compose([
    tt.RandomCrop(size = (128, 128)),
    tt.RandomHorizontalFlip()
])
transformed_images = transforms(dummy_images)
print(transformed_images.size())

# As you can see, Compose offers us the option of sequentially
# executing multiple transformations in a single line of code.
# We also get the option of using randomized augmentations,
# where the randomization is already done for us.

# In practice, either style of writing transformations is fine.
# In fact, they are equivalent, as Compose calls the functional
# versions of the transforms under the hood. In the case of
# randomized augmentations, the class handles all the randomizing
# and then calls the functional transform with the random inputs.
```

#### Chapter 3.2 - Regularization Techniques

While there are multiple definitions or guidelines for what regularization is supposed to do (see lectures), in terms of practical concerns, all regularization techniques have the same aim, expressed through different means: Improving some aspect of the performance of your deep learning models. We differentiate them, broadly, from data augmentations, because regularization techniques generally concern themselves with the learning process, e.g. loss function modifications, learning rate optimizations, temporary model modifications, etc., and *not* the underlying data in our model training.

There are a number of different strategies, far too many to list all of them here, but a few particularly successful ones have made it into common use - so much so that they are more prevalent than regularization-less, "vanilla" deep learning. These fall into different groups, briefly discussed below.

#### Additional Loss Components

The loss function for any given modern optimization task is typically continuous and not always smooth everywhere. As a consequence, there are many different parameter configurations in a model that result in the same train-time loss. Not all of these express the same behavior during training or testing, however. When we modify our loss to penalize certain training behaviors, we allow the training process to select for models and parameters that give models that generalize better, converge to a solution faster, etc.,  despite often expressing the same training loss. Let's look at some examples that should be familiar from the lecture:

**L1 Loss** - Often also called LASSO, L1 Loss is a penalty term added to the normal loss during training, which is defined as:
$L_{LASSO} = \sum_{p=1}^{P} |\Theta_{P}|$. Growing linearly with parameter magnitude, we penalize the model. This, in turn, forces the model to use fewer and smaller weights - relying on more weights than it needs, and thus probably overfitting, is disincentivized. Similarly, we just forced our model to stick to weights near zero, which we already know is generally a preferable area for model activations to stick to.

**L2 Loss** - L2 Loss is the more popular cousin of the L1 Loss, which does approximately the same thing, except the penalty is equal to the sum over all squared parameter magnitudes: $L_{LASSO} = \sum_{p=1}^{P} |\Theta_{P}|^2$ The reasoning behind it is similar, but it has seen far more practical adoption.

**Weight Decay** - An operation that effectively performs the same duty, weight decay reduces the magnitude of weights after each backward pass, for example by subtracting a small constant or multiplying with a factor. In essence, this eliminates parameters which are rarely "used" and were thus likely involved in overfitting on a small amount of data anyway. Parameters that are regularly updated (and therefore probably useful), will always remain near their optimal value despite weight decay. Interestingly, Weight Decay is mathematically equivalent to L2 Loss in terms of net parameter updates.

#### Training Strategies
**Early Stopping**: For early stopping, we monitor the performance of the model on a validation set and stop training when the performance stops improving. This ensures that the model does not continue to train on the training data and potentially overfit, while also saving computational resources.

**Dropout**: Dropout is a regularization technique where, during training, a random subset of neurons in a layer is "dropped out" (set to zero) for each forward pass. This prevents the network from relying too heavily on specific neurons and encourages it to learn more robust and generalized features. During inference, all neurons are used, but their outputs are scaled to account for the dropout during training.

**Learning Rate Scheduling (LR Scheduling)**: Learning rate scheduling involves dynamically adjusting the learning rate during training. A high learning rate at the start helps the model converge quickly, while a lower learning rate later allows for fine-tuning. Common strategies include step decay, exponential decay, and cosine annealing. Proper learning rate scheduling can lead to faster convergence and better generalization.

```python



class LiTS_Dataset(Dataset):

    """
    For our sample solution, we go for the easier variant.

    In this specific dataset, we don't load the images until we need them - for a
    short training, or limited resources, this is good behavior. If you have the
    necessary RAM to pre-load all of your data, you don't have to load the data
    multiple times, and save compute costs in the long run. The downside is that
    when you are trying to debug, you wait for ages every time, and if you simply
    do not have the compute resources, you can't even do it.
    """

    def __init__(self, csv: str, mode: str):

        self.csv = csv
        self.data = pd.read_csv(self.csv)
        self.mode = mode
        assert mode in ["train", "val", "test"] # has to be train, val, or test data - if not, assert throws an error

    def __len__(self):

        return len(self.data)

    def __getitem__(self, idx):

        file = self.data.loc[idx, "filename"]
        with PIL.Image.open(f"./Clean_LiTS/{self.mode}/{file}") as f:
            f = f.convert("L")
            image = ttf.pil_to_tensor(f)

        image = image.to(dtype = torch.float32)
        image -= torch.min(image)
        image /= torch.max(image)

        liver_visible = self.data.loc[idx, "liver_visible"]
        lesion_visible = self.data.loc[idx, "lesion_visible"]
        # Note that targets must have the data type torch.long - a 64-bit integer,
        # unlike the image tensor, which is usually a 32-bit float, the default
        # dtype for tensors when none is given
        if lesion_visible and liver_visible:
            target = torch.tensor(2, dtype = torch.long)
        elif not lesion_visible and liver_visible:
            target = torch.tensor(1, dtype = torch.long)
        elif not lesion_visible and not liver_visible:
            target = torch.tensor(0, dtype = torch.long)
        else:
            print(
                idx,
                lesion_visible,
                liver_visible,
                self.data.loc[idx, "liver_visible"],
                self.data.loc[idx, "lesion_visible"],
                self.data.loc[idx, "filename"]
                )
            raise ValueError("Invalid target")

        return image, target

train_dataset = LiTS_Dataset(csv = "./Clean_LiTS/train_classes.csv", mode="train")
val_dataset = LiTS_Dataset(csv = "./Clean_LiTS/val_classes.csv", mode="val")
test_dataset = LiTS_Dataset(csv = "./Clean_LiTS/test_classes.csv", mode="test")

batch_size = 16

train_dataloader = DataLoader(
    dataset = train_dataset,
    batch_size = batch_size,
    shuffle = True,
    drop_last = True,
    num_workers = 24
)

val_dataloader = DataLoader(
    dataset = val_dataset,
    batch_size = batch_size,
    num_workers = 24,
    shuffle = False,
    drop_last = True
)

test_dataloader = DataLoader(
    dataset = test_dataset,
    batch_size = batch_size,
    shuffle = False,
    drop_last = True,
    num_workers = 24
)
```

```python
device = ("cuda" if torch.cuda.is_available() else "cpu")

#class Really_Good_Model_forLIST(nn.Module):
#    """
#    Ein kleines CNN, das am Ende 3 Klassen ausgibt.
#    """
#
#    def __init__(self, input_shape):
#        super().__init__()


class YourModel(nn.Module):
    def __init__(self, in_channels=1, out_classes=3):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels, 12, kernel_size=3)
        self.relu1 = nn.ReLU()
        self.conv2 = nn.Conv2d(12, 12, kernel_size=3)
        self.relu2 = nn.ReLU()
        self.global_pool = nn.AdaptiveAvgPool2d((1,1))  # auf 1x1 reduzieren
        self.fc = nn.Linear(12, out_classes)  # 12 Kanle -> out_classes

    def forward(self, x):
        x = self.relu1(self.conv1(x))
        x = self.relu2(self.conv2(x))
        x = self.global_pool(x)
        x = torch.flatten(x, 1)
        x = self.fc(x)
        return x

```

```python
# Now for the training loop
loss_criterion = torch.nn.CrossEntropyLoss()

def train_loop(your_model):
    """
    This function runs your train loop and returns the trained model.
    """

    model = your_model(in_channels = 1, out_classes = 3)
    model = model.to(device = device)

    loss_criterion = torch.nn.CrossEntropyLoss()

    optimizer = torch.optim.Adam(params = model.parameters(), lr = 1e-4)

    num_epochs = 15

    for epoch in range(num_epochs):

        for step, (data, targets) in enumerate(train_dataloader):

            optimizer.zero_grad()
            data, targets = data.to(device), targets.to(device)
            predictions = model(data)
            loss = loss_criterion(predictions, targets)

            if step % 50 == 0:
                # This uses the length of the current set - "train"
                print(f"Epoch [{epoch+1}/{num_epochs}]\t Step [{step+1}/{len(train_dataloader.dataset)//batch_size}]\t Loss: {loss.item():.4f}")

            loss.backward()
            optimizer.step()

        # Validate every 2 epochs
        if epoch % 2 == 0:

            # Validation mode on
            model.eval()

            # Don't track gradients for validation
            with torch.no_grad():

                hits = 0
                losses = []
                batch_sizes = []

                for step, (data, targets) in enumerate(val_dataloader):

                    data, targets = data.to(device), targets.to(device)
                    predictions = model(data)
                    loss = loss_criterion(predictions, targets)
                    losses.append(loss.item())
                    batch_sizes.append(data.size()[0])

                    class_predictions = torch.argmax(predictions, dim = 1).flatten()
                    hits = hits + sum([1 if cp == t else 0 for cp, t in zip(class_predictions, targets)])

                accuracy = hits / len(val_dataloader.dataset)
                avg_loss = sum([l * bs for l, bs in zip(losses, batch_sizes)]) / sum(batch_sizes)
                print(f"Epoch: {epoch+1},\t Validation Loss: {avg_loss:.4f},\t Accuracy: {accuracy:.4f}")

            # After we are done validating, let's not forget to go back to storing gradients.
            model.train()

    return model
```

```python
trained_model = train_loop(YourModel)
```

```python
# And finally, test the model

trained_model.eval()

with torch.no_grad():

    hits = 0
    losses = []
    batch_sizes = []

    for step, (data, targets) in enumerate(test_dataloader):

        data, targets = data.to(device), targets.to(device)
        predictions = trained_model(data)
        loss = loss_criterion(predictions, targets)
        losses.append(loss.item())
        batch_sizes.append(data.size()[0])
        class_predictions = torch.argmax(predictions, dim = 1).flatten()
        hits = hits + sum([1 if cp == t else 0 for cp, t in zip(class_predictions, targets)])

    accuracy = hits / len(test_dataloader.dataset)
    avg_loss = sum([l * bs for l, bs in zip(losses, batch_sizes)]) / sum(batch_sizes)
    print(f"Test Loss: {avg_loss:.4f},\t Accuracy: {accuracy:.4f}")
```

#### Chapter 3.3 - Batch Normalization

Batch Normalization is a technique to improve the training of deep neural networks by normalizing the inputs to each layer. It was introduced to address the problem of internal covariate shift, which refers to the change in the distribution of layer inputs during training as the parameters of the previous layers change.

**Intuition:**
The idea behind Batch Normalization is to normalize the inputs to each layer so that they have a mean of 0 and a standard deviation of 1. This ensures that the inputs to each layer are on a similar scale, which helps the network learn faster and more effectively. By normalizing the inputs, Batch Normalization reduces the sensitivity of the network to the initialization of weights and allows for the use of higher learning rates.

**How it works:**
1. For each mini-batch, Batch Normalization computes the mean and variance of the inputs.
2. The inputs are then normalized using these statistics:
    $\hat{x} = \frac{x - \mu}{\sqrt{\sigma^2 + \epsilon}}$
    where $\mu$ is the mean, $\sigma^2$ is the variance, and $\epsilon$ is a small constant added for numerical stability.
3. To allow the network to learn the optimal scale and shift for the normalized inputs, two learnable parameters, $\gamma$ (scale) and $\beta$ (shift), are introduced:
    $y = \gamma \hat{x} + \beta$

**Problems it solves:**
1. **Internal Covariate Shift:** By normalizing the inputs to each layer, Batch Normalization reduces the changes in the distribution of layer inputs during training, making the optimization process more stable.
2. **Faster Training:** Normalized inputs allow for the use of higher learning rates, leading to faster convergence.
3. **Regularization Effect:** Batch Normalization introduces some noise due to the mini-batch statistics, which acts as a form of regularization and reduces the need for other regularization techniques like Dropout.

```python

class AlexNet(nn.Module):
    def __init__(self, num_classes=3, in_channels=1):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(in_channels, 64, kernel_size=11, stride=4, padding=2),
            nn.ReLU(),
            nn.MaxPool2d(3, 2),

            nn.Conv2d(64, 192, 5, padding=2),
            nn.ReLU(),
            nn.MaxPool2d(3, 2),

            nn.Conv2d(192, 384, 3, padding=1),
            nn.ReLU(),

            nn.Conv2d(384, 256, 3, padding=1),
            nn.ReLU(),

            nn.Conv2d(256, 256, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(3, 2)
        )

        # compute output size automatically
        with torch.no_grad():
            dummy = torch.zeros(1, in_channels, 224, 224)
            out = self.features(dummy)
            self.flatten_dim = out.shape[1] * out.shape[2] * out.shape[3]


        self.classifier = nn.Sequential(
            nn.Dropout(),
            nn.Linear(self.flatten_dim, 4096),
            nn.ReLU(),

            nn.Dropout(),
            nn.Linear(4096, 4096),
            nn.ReLU(),

            nn.Linear(4096, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)   # uses flatten_dim automatically
        return self.classifier(x)
```

## ResNet-50: Architektur, Training und Auswertung

Die folgenden Zellen definieren einen vollstndig kommentierten ResNet-50 mit Skip Connections (Bottleneck-Blcke), erweitern das Trainingsskript um Datenaugmentation, Regularisierung (Weight Decay & Dropout), Validierungsprotokollierung mit Early Stopping sowie eine Auswertung auf dem Testset inkl.
```python
# === berblick ===  # Kommentarblock: Diese Zelle definiert das komplette ResNet-50 (Bottleneck-Blcke, Skip Connections),
# einschlielich Hilfsfunktionen, Gewichtsinitialisierung und einer Instanz fr unser 3-Klassen-Problem.
# Alle Schritte sind deutsch kommentiert, damit klar bleibt, welche Bausteine welche Aufgabe erfllen.
# ResNet-50 Definition mit Bottleneck-Blcken und Skip Connections  # kommentiert jede Zeile deutsch
import copy  # Hilfsmodul fr tiefe Kopien (Early Stopping)
import numpy as np  # Numerische Arrays fr Auswertung
import matplotlib.pyplot as plt  # Plotten von Trainingskurven
from typing import List  # Typ-Hilfe fr Listen von Modulen
import torchvision.transforms as tt  # Torchvision-Transforms fr Augmentation
from torch.utils.data import DataLoader  # Dataloader-Typ fr Typ-Hinweise
from tqdm.auto import tqdm  # Fortschrittsbalken fr Notebooks

# Hilfsfunktion: 3x3-Conv mit Padding, um Hhe/Breite zu erhalten und optional zu striden.
def conv3x3(in_channels: int, out_channels: int, stride: int = 1) -> nn.Conv2d:  # 3x3 Faltung mit Padding
    return nn.Conv2d(in_channels, out_channels, kernel_size=3, stride=stride, padding=1, bias=False)  # Kernel ohne Bias

# Hilfsfunktion: 1x1-Conv fr Kanalreduktion/-expansion und Stride-Anpassung im Skip-Pfad.
def conv1x1(in_channels: int, out_channels: int, stride: int = 1) -> nn.Conv2d:  # 1x1 Faltung zur Kanalreduktion/Expansion
    return nn.Conv2d(in_channels, out_channels, kernel_size=1, stride=stride, bias=False)  # Bias weggelassen wegen BatchNorm

# Bottleneck-Block: Drei Convs + Skip Connection, reduziert erst Kanle, extrahiert Features, erweitert zurck.
class Bottleneck(nn.Module):  # Bottleneck-Block mit Expansion und Skip Connection
    expansion = 4  # Faktor zur Erweiterung der Kanle am Blockende

    def __init__(self, in_channels: int, bottleneck_channels: int, stride: int = 1, downsample: nn.Module | None = None):  # Konstruktor
        super().__init__()  # Basiskonstruktor
        self.conv1 = conv1x1(in_channels, bottleneck_channels)  # 1x1 reduziert Kanle
        self.bn1 = nn.BatchNorm2d(bottleneck_channels)  # BatchNorm stabilisiert erste Stufe
        self.conv2 = conv3x3(bottleneck_channels, bottleneck_channels, stride=stride)  # 3x3 extrahiert Features
        self.bn2 = nn.BatchNorm2d(bottleneck_channels)  # BatchNorm nach 3x3
        self.conv3 = conv1x1(bottleneck_channels, bottleneck_channels * self.expansion)  # 1x1 expandiert zurck
        self.bn3 = nn.BatchNorm2d(bottleneck_channels * self.expansion)  # BatchNorm nach Expansion
        self.relu = nn.ReLU(inplace=True)  # ReLU aktiviert inplace zur Speicherersparnis
        self.downsample = downsample  # Optionaler Pfad zum Angleichen von Gre/Kanlen

    def forward(self, x: torch.Tensor) -> torch.Tensor:  # Vorwrtsdurchlauf des Blocks
        identity = x  # Speichere Eingang fr Skip Connection

        out = self.conv1(x)  # 1x1 Reduktion
        out = self.bn1(out)  # Normalisierung
        out = self.relu(out)  # Aktivierung

        out = self.conv2(out)  # 3x3 Faltung (ggf. mit Stride)
        out = self.bn2(out)  # Normalisierung
        out = self.relu(out)  # Aktivierung

        out = self.conv3(out)  # 1x1 Expansion
        out = self.bn3(out)  # Normalisierung

        if self.downsample is not None:  # Falls Dimensionen nicht passen
            identity = self.downsample(x)  # Passe Skip-Pfad an

        out += identity  # Fge Skip Connection hinzu
        out = self.relu(out)  # Aktivierung nach Summation
        return out  # Ausgabe des Blocks

# ResNet50: Stapelt vier Stages der Bottlenecks, kombiniert Global Pooling, Dropout und finale Linearschicht.
class ResNet50(nn.Module):  # ResNet-50 Definition
    def __init__(self, num_classes: int = 3, in_channels: int = 1, dropout_p: float = 0.2):  # Konstruktor
        super().__init__()  # Basiskonstruktor
        self.in_channels = 64  # Startkanle nach erstem Conv
        self.conv1 = nn.Conv2d(in_channels, self.in_channels, kernel_size=7, stride=2, padding=3, bias=False)  # Erster 7x7 Conv
        self.bn1 = nn.BatchNorm2d(self.in_channels)  # Normalisierung nach Conv1
        self.relu = nn.ReLU(inplace=True)  # ReLU Aktivierung
        self.maxpool = nn.MaxPool2d(kernel_size=3, stride=2, padding=1)  # Reduziert Auflsung frh

        self.layer1 = self._make_layer(64, blocks=3)  # Erste Stage mit 3 Bottlenecks
        self.layer2 = self._make_layer(128, blocks=4, stride=2)  # Zweite Stage mit Downsampling
        self.layer3 = self._make_layer(256, blocks=6, stride=2)  # Dritte Stage (grter Block)
        self.layer4 = self._make_layer(512, blocks=3, stride=2)  # Vierte Stage

        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))  # Globale Mittelung auf 1x1
        self.dropout = nn.Dropout(dropout_p)  # Dropout als Regularisierung
        self.fc = nn.Linear(512 * Bottleneck.expansion, num_classes)  # Finale Klassifikation

        for m in self.modules():  # Gewichtsinitialisierung
            if isinstance(m, nn.Conv2d):  # Fr Convs
                nn.init.kaiming_normal_(m.weight, mode="fan_out", nonlinearity="relu")  # Kaiming Init
            elif isinstance(m, nn.BatchNorm2d):  # Fr BatchNorms
                nn.init.constant_(m.weight, 1)  # Gamma = 1
                nn.init.constant_(m.bias, 0)  # Beta = 0

    def _make_layer(self, bottleneck_channels: int, blocks: int, stride: int = 1) -> nn.Sequential:  # Baut eine ResNet-Stage
        downsample = None  # Startwert fr Downsampling-Pfad
        if stride != 1 or self.in_channels != bottleneck_channels * Bottleneck.expansion:  # Prfe Dim-Abgleich
            downsample = nn.Sequential(  # Passe Skip an
                conv1x1(self.in_channels, bottleneck_channels * Bottleneck.expansion, stride=stride),  # 1x1 mit Stride
                nn.BatchNorm2d(bottleneck_channels * Bottleneck.expansion),  # Normalisierung
            )  # Ende Downsample-Sequenz

        layers: List[nn.Module] = []  # Liste fr Blcke
        layers.append(Bottleneck(self.in_channels, bottleneck_channels, stride=stride, downsample=downsample))  # Erster Block mit evtl. Downsample
        self.in_channels = bottleneck_channels * Bottleneck.expansion  # Update aktuelle Kanalzahl
        for _ in range(1, blocks):  # Restliche Blcke
            layers.append(Bottleneck(self.in_channels, bottleneck_channels))  # Standard-Bottleneck ohne Downsample

        return nn.Sequential(*layers)  # Gebe Stage zurck

    def forward(self, x: torch.Tensor) -> torch.Tensor:  # Vorwrtsdurchlauf des Netzes
        x = self.conv1(x)  # Erste Faltung
        x = self.bn1(x)  # Normalisierung
        x = self.relu(x)  # Aktivierung
        x = self.maxpool(x)  # Pooling

        x = self.layer1(x)  # Stage 1
        x = self.layer2(x)  # Stage 2
        x = self.layer3(x)  # Stage 3
        x = self.layer4(x)  # Stage 4

        x = self.avgpool(x)  # Globale Mittelung
        x = torch.flatten(x, 1)  # Vektor machen
        x = self.dropout(x)  # Dropout anwenden
        x = self.fc(x)  # Logits berechnen
        return x  # Ausgabe zurckgeben

# Modellinstanz
resnet50 = ResNet50(num_classes=3, in_channels=1, dropout_p=0.3).to(device)  # Erzeuge ResNet-50 und schiebe auf Gert
```

```python
# === berblick Training ===  # Kommentarblock: Diese Zelle regelt die Augmentierungen, das Training mit Early Stopping, Logging,
# LR-Scheduling, sowie das Plotten der Verlufe. Zustzlich werden Weight Decay und Dropout als Regularisierung eingesetzt.
# Trainings-Setup mit Datenaugmentation, Regularisierung, Validierungs-Logging und Early Stopping  # berschrift
train_transforms_resnet = tt.Compose([  # Trainings-Augs definieren
    tt.Resize((256, 256)),  # Grere Ausgangsgre fr zufllige Crops
    tt.RandomResizedCrop(224, scale=(0.9, 1.0)),  # Zuflliger Ausschnitt auf 224x224
    tt.RandomHorizontalFlip(p=0.5),  # Spiegelung zur Robustheit
    tt.RandomRotation(degrees=10),  # Kleine Rotation
    tt.RandomAffine(degrees=0, translate=(0.05, 0.05)),  # Kleine Verschiebung
])  # Ende Compose

val_transforms_resnet = tt.Compose([  # Val-/Test-Transform
    tt.Resize((224, 224)),  # Nur Resize ohne Augmentation
])  # Ende Compose

# Eval-Funktion: fhrt einen kompletten Durchlauf ber Loader durch und berechnet Loss/Accuracy ohne Gradienten.
def evaluate_on_loader(model: nn.Module, loader: DataLoader, transform, criterion):  # Eval-Funktion
    model.eval()  # Schalte in Eval-Modus
    total, correct, running_loss = 0, 0, 0.0  # Akkumulatoren
    with torch.no_grad():  # Kein Gradienten-Tracking
        for data, targets in loader:  # Batches iterieren
            data = torch.stack([transform(img) for img in data]).to(device)  # Transformiere und auf Device
            targets = targets.to(device)  # Targets auf Device
            preds = model(data)  # Vorhersage
            loss = criterion(preds, targets)  # Loss berechnen
            running_loss += loss.item() * data.size(0)  # Loss aufsummieren
            correct += (preds.argmax(dim=1) == targets).sum().item()  # Treffer zhlen
            total += data.size(0)  # Batchgre aufsummieren
    avg_loss = running_loss / max(total, 1)  # Durchschnittlicher Loss
    acc = correct / max(total, 1)  # Durchschnittliche Accuracy
    return avg_loss, acc  # Rckgabe

# Trainingsfunktion: fhrt Epochen durch, nutzt Scheduler, Early Stopping und speichert bestes Modell.
def train_resnet(model: nn.Module, epochs: int = 2, patience: int = 4, lr: float = 1e-3, weight_decay: float = 1e-4):  # Training
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)  # Adam mit Weight Decay
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode="min", factor=0.5, patience=2)  # LR-Scheduler
    criterion = nn.CrossEntropyLoss()  # Klassifikations-Loss

    history = {"train_loss": [], "val_loss": [], "val_acc": []}  # Logging
    best_state = copy.deepcopy(model.state_dict())  # Bestes Modell merken
    best_val = float("inf")  # Bestes Val-Loss initialisieren
    wait = 0  # Patience-Zhler
    pbar = tqdm(total=epochs * len(train_dataloader), desc="Training", leave=True)  # Eine Gesamt-tqdm ber alle Schritte
    pending_update = 0  # Zhler fr verzgertes Update (alle 250 Schritte)

    for epoch in range(epochs):  # Epochen-Schleife
        model.train()  # Train-Modus
        epoch_loss = 0.0  # Epochen-Loss
        total = 0  # Anzahl Samples

        for step, (data, targets) in enumerate(train_dataloader):  # Batches trainieren
            data = torch.stack([train_transforms_resnet(img) for img in data]).to(device)  # Augmentiere Batch
            targets = targets.to(device)  # Targets auf Device

            optimizer.zero_grad()  # Gradienten nullen
            preds = model(data)  # Vorwrts-Pass
            loss = criterion(preds, targets)  # Loss berechnen
            loss.backward()  # Backprop
            optimizer.step()  # Gewichte updaten

            batch_size = data.size(0)  # Aktuelle Batchgre
            epoch_loss += loss.item() * batch_size  # Loss aufsummieren
            total += batch_size  # Samples zhlen
            pending_update += 1  # Schritt zhlen
            if pending_update >= 250:  # Nur alle 250 Schritte updaten
                pbar.update(pending_update)  # Fortschritt erhhen
                pending_update = 0  # Zhler zurcksetzen

        avg_train_loss = epoch_loss / max(total, 1)  # Durchschnittlicher Train-Loss
        val_loss, val_acc = evaluate_on_loader(model, val_dataloader, val_transforms_resnet, criterion)  # Val-Auswertung

        history["train_loss"].append(avg_train_loss)  # Train-Loss loggen
        history["val_loss"].append(val_loss)  # Val-Loss loggen
        history["val_acc"].append(val_acc)  # Val-Acc loggen

        scheduler.step(val_loss)  # Scheduler mit Val-Loss updaten

        pbar.set_postfix({  # Zeige aktuelle Kennzahlen in der Leiste
            "epoch": f"{epoch+1}/{epochs}",
            "train_loss": f"{avg_train_loss:.4f}",
            "val_loss": f"{val_loss:.4f}",
            "val_acc": f"{val_acc:.4f}",
            "lr": f"{optimizer.param_groups[0]['lr']:.2e}"
        })

        if val_loss < best_val - 1e-4:  # Verbesserung?
            best_val = val_loss  # Bestes Loss updaten
            best_state = copy.deepcopy(model.state_dict())  # Modell sichern
            wait = 0  # Patience reset
        else:  # Keine Verbesserung
            wait += 1  # Patience erhhen
            if wait >= patience:  # Patience erschpft?
                pbar.write(f"Frhes Stoppen nach {epoch+1} Epochen (bestes Val-Loss: {best_val:.4f}).")  # Info ausgeben
                break  # Training beenden

    if pending_update > 0:  # Restliche Schritte aktualisieren
        pbar.update(pending_update)  # Fortschritt schlieen
    pbar.close()  # Leiste schlieen
    model.load_state_dict(best_state)  # Bestes Modell laden
    return history  # Trainingshistorie zurckgeben

# Training ausfhren
history_resnet = train_resnet(resnet50, epochs=50, patience=5, lr=1e-3, weight_decay=1e-4)  # Starte Training

# Trainingsverlauf plotten (nach Trainingsende)
epochs_range = range(1, len(history_resnet["train_loss"]) + 1)  # Epochenachse
plt.figure(figsize=(10, 4))  # Figure anlegen
plt.subplot(1, 2, 1)  # Erstes Panel
plt.plot(epochs_range, history_resnet["train_loss"], label="Train-Loss")  # Train-Loss Plot
plt.plot(epochs_range, history_resnet["val_loss"], label="Val-Loss")  # Val-Loss Plot
plt.xlabel("Epoche")  # Achsenbeschriftung
plt.ylabel("Loss")  # Achsenbeschriftung
plt.legend()  # Legende
plt.title("Loss-Verlauf mit Early Stopping")  # Titel

plt.subplot(1, 2, 2)  # Zweites Panel
plt.plot(epochs_range, history_resnet["val_acc"], label="Val-Accuracy", color="tab:green")  # Val-Acc Plot
plt.xlabel("Epoche")  # Achsenbeschriftung
plt.ylabel("Accuracy")  # Achsenbeschriftung
plt.title("Validierungsgenauigkeit je Epoche")  # Titel
plt.legend()  # Legende
plt.tight_layout()  # Layout anpassen
plt.show()  # Plot anzeigen
```

```python
# === berblick Test ===  # Kommentarblock: Diese Zelle wertet das trainierte ResNet-50 auf dem Testset aus,
# berechnet Loss und Accuracy, zeigt eine Konfusionsmatrix und Precision-Recall-Kurven pro Klasse, und gibt bei Fehlern
# eine klare Meldung aus.
# Test-Auswertung mit Konfusionsmatrix und Precision-Recall-Kurven  # berschrift
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_recall_curve, PrecisionRecallDisplay  # Metrik-Funktionen

try:  # Fehler abfangen
    resnet50.eval()  # Modell in Eval-Modus
    criterion = nn.CrossEntropyLoss()  # Loss fr Konsistenz
    all_preds, all_targets, all_probs = [], [], []  # Sammler fr Auswertung
    hits, losses, batch_sizes = 0, [], []  # Zhler und Losslisten

    with torch.no_grad():  # Kein Gradienten-Tracking
        for data, targets in test_dataloader:  # Testdaten iterieren
            data = torch.stack([val_transforms_resnet(img) for img in data]).to(device)  # Transformiere/Device
            targets = targets.to(device)  # Targets auf Device
            preds = resnet50(data)  # Vorhersagen
            prob = torch.softmax(preds, dim=1)  # Softmax fr Wahrscheinlichkeiten
            all_probs.append(prob.cpu().numpy())  # Speichern
            loss = criterion(preds, targets)  # Test-Loss
            losses.append(loss.item())  # Loss sammeln
            batch_sizes.append(data.size(0))  # Batchgre sammeln
            pred_labels = preds.argmax(dim=1)  # Klassen whlen
            hits += (pred_labels == targets).sum().item()  # Treffer zhlen
            all_preds.extend(pred_labels.cpu().tolist())  # Vorhersagen sichern
            all_targets.extend(targets.cpu().tolist())  # Ziele sichern

    test_acc = hits / len(test_dataloader.dataset)  # Gesamt-Accuracy
    test_loss = sum(l * bs for l, bs in zip(losses, batch_sizes)) / sum(batch_sizes)  # Gewichteter Loss
    print(f"ResNet50 Test Loss: {test_loss:.4f}\tAccuracy: {test_acc:.4f}")  # Ausgabe

    cm = confusion_matrix(all_targets, all_preds)  # Konfusionsmatrix
    ConfusionMatrixDisplay(confusion_matrix=cm).plot(cmap=plt.cm.Blues)  # Plot Matrix
    plt.title("Konfusionsmatrix ResNet-50")  # Titel
    plt.show()  # Anzeigen

    num_classes = max(all_targets) + 1  # Klassenanzahl bestimmen
    y_true = torch.nn.functional.one_hot(torch.tensor(all_targets), num_classes=num_classes).numpy()  # One-Hot Targets
    y_scores = np.concatenate(all_probs, axis=0)  # Alle Softmax-Scores stapeln

    for i in range(num_classes):  # Fr jede Klasse PR-Kurve
        precision, recall, _ = precision_recall_curve(y_true[:, i], y_scores[:, i])  # PR-Kurve berechnen
        PrecisionRecallDisplay(precision=precision, recall=recall).plot()  # Plotten
        plt.title(f"Precision-Recall (Klasse {i})")  # Titel setzen
        plt.show()  # Anzeigen

except Exception as e:  # Fehlerfall
    print(f"Fehler bei der ResNet-Auswertung: {e}")  # Klare Fehlermeldung
```

```python
# === berblick Pretrained Training ===  # Kommentarblock: Diese Zelle ldt ein vortrainiertes ResNet-50, definiert passende ImageNet-Transforms, trainiert mit Early Stopping und loggt Loss/Accuracy.
# Pretrained ResNet-50 mit ImageNet-Normalisierung, Augmentation, TQDM-Logging und Validierungsprotokollierung
from torchvision.models import resnet50, ResNet50_Weights  # Pretrained ResNet50-Architektur und Gewichte importieren
imagenet_weights = ResNet50_Weights.IMAGENET1K_V2  # Gewhlte ImageNet-Gewichte laden
imagenet_mean = (0.485, 0.456, 0.406)  # Feste ImageNet-Mittelwerte (Fallback falls Metadaten fehlen)
imagenet_std = (0.229, 0.224, 0.225)  # Feste ImageNet-Std (Fallback falls Metadaten fehlen)
train_transforms_pretrained = tt.Compose([  # Trainings-Transforms fr das vortrainierte Netz
    tt.Resize((256, 256)),  # Grere Basisauflsung vor dem Crop
    tt.RandomResizedCrop(224, scale=(0.9, 1.0)),  # Zuflliger Zuschnitt auf 224x224
    tt.RandomHorizontalFlip(p=0.5),  # Spiegelung zur Datenaugmentation
    tt.RandomRotation(degrees=10),  # Kleine Rotationen als Robustheit
    tt.RandomAffine(degrees=0, translate=(0.05, 0.05)),  # Leichte Verschiebungen
    tt.Lambda(lambda x: x.repeat(3, 1, 1)),  # Graukanal auf 3 Kanle duplizieren
    tt.Normalize(mean=imagenet_mean, std=imagenet_std),  # ImageNet-Normalisierung anwenden
])  # Ende Compose fr Training
val_transforms_pretrained = tt.Compose([  # Val/Test-Transforms fr das vortrainierte Netz
    tt.Resize((224, 224)),  # Resize auf 224x224 ohne Augmentation
    tt.Lambda(lambda x: x.repeat(3, 1, 1)),  # 3-Kanal-Kopie fr pretrained Eingang
    tt.Normalize(mean=imagenet_mean, std=imagenet_std),  # ImageNet-Normalisierung
])  # Ende Compose fr Val/Test
pretrained_resnet = resnet50(weights=imagenet_weights)  # ResNet-50 mit ImageNet-Gewichten laden
pretrained_resnet.fc = nn.Linear(pretrained_resnet.fc.in_features, 3)  # Finale Schicht auf 3 Klassen anpassen
pretrained_resnet = pretrained_resnet.to(device)  # Modell auf GPU/CPU verschieben
# Trainingsfunktion: nutzt TQDM (ein Balken ber alle Epochen), Early Stopping, Scheduler und Logging.
def train_pretrained_resnet(model: nn.Module, epochs: int = 20, patience: int = 4, lr: float = 5e-4, weight_decay: float = 1e-4):  # Training definieren
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)  # Adam-Optimierer mit Weight Decay
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)  # LR-Scheduler fr Plateaus
    criterion = nn.CrossEntropyLoss()  # Klassifikations-Loss definieren
    history = {'train_loss': [], 'val_loss': [], 'val_acc': []}  # Verlaufslogging initialisieren
    best_state = copy.deepcopy(model.state_dict())  # Bestes Modell zwischenspeichern
    best_val = float('inf')  # Bestes Val-Loss initialisieren
    wait = 0  # Patience-Zhler starten
    pbar = tqdm(total=epochs * len(train_dataloader), desc='Pretrained Training', leave=True)  # Gesamt-tqdm ber alle Epochen/Batches
    pending_update = 0  # Schritte bis zum nchsten tqdm-Update (alle 250)
    for epoch in range(epochs):  # Epochen-Schleife starten
        model.train()  # Train-Modus aktivieren
        epoch_loss = 0.0  # Akkumulator fr Epochen-Loss
        total = 0  # Zhler fr Samples
        for step, (data, targets) in enumerate(train_dataloader):  # Batches der Trainingsdaten durchlaufen
            data = torch.stack([train_transforms_pretrained(img) for img in data]).to(device)  # Augmentieren und auf Device legen
            targets = targets.to(device)  # Targets auf Device legen
            optimizer.zero_grad()  # Gradienten zurcksetzen
            preds = model(data)  # Vorwrts-Pass berechnen
            loss = criterion(preds, targets)  # Loss ausrechnen
            loss.backward()  # Backpropagation ausfhren
            optimizer.step()  # Optimizer-Schritt machen
            batch_size = data.size(0)  # Aktuelle Batchgre ermitteln
            epoch_loss += loss.item() * batch_size  # Loss gewichtet aufsummieren
            total += batch_size  # Sample-Zahl erhhen
            pending_update += 1  # Schrittzhler fr tqdm erhhen
            if pending_update >= 250:  # Nur alle 250 Schritte die Leiste updaten
                pbar.update(pending_update)  # Fortschritt aktualisieren
                pending_update = 0  # Zhler zurcksetzen
        avg_train_loss = epoch_loss / max(total, 1)  # Durchschnittlichen Train-Loss berechnen
        val_loss, val_acc = evaluate_on_loader(model, val_dataloader, val_transforms_pretrained, criterion)  # Validierung ausfhren
        history['train_loss'].append(avg_train_loss)  # Train-Loss loggen
        history['val_loss'].append(val_loss)  # Val-Loss loggen
        history['val_acc'].append(val_acc)  # Val-Accuracy loggen
        scheduler.step(val_loss)  # Scheduler mit Val-Loss fttern
        pbar.set_postfix({  # Kennzahlen in tqdm anzeigen
            'epoch': f'{epoch+1}/{epochs}',  # Aktuelle Epoche
            'train_loss': f'{avg_train_loss:.4f}',  # Train-Loss anzeigen
            'val_loss': f'{val_loss:.4f}',  # Val-Loss anzeigen
            'val_acc': f'{val_acc:.4f}',  # Val-Accuracy anzeigen
            'lr': f"{optimizer.param_groups[0]['lr']:.2e}"  # Aktuelle Learning Rate
        })  # Ende Postfix
        if val_loss < best_val - 1e-4:  # Prfen, ob Val-Loss besser ist
            best_val = val_loss  # Bestes Val-Loss aktualisieren
            best_state = copy.deepcopy(model.state_dict())  # Bestes Modell speichern
            wait = 0  # Patience zurcksetzen
        else:  # Keine Verbesserung
            wait += 1  # Patience erhhen
            if wait >= patience:  # Patience ausgeschpft?
                pbar.write(f'Frhes Stoppen nach {epoch+1} Epochen (bestes Val-Loss: {best_val:.4f}).')  # Info ausgeben
                break  # Training abbrechen
    if pending_update > 0:  # Letzte Schritte in tqdm nachtragen
        pbar.update(pending_update)  # Fortschritt abschlieen
    pbar.close()  # Leiste schlieen
    model.load_state_dict(best_state)  # Bestes Modell laden
    return history  # Historie zurckgeben
history_resnet_pretrained = train_pretrained_resnet(pretrained_resnet, epochs=25, patience=5, lr=5e-4, weight_decay=1e-4)  # Training starten
epochs_range_pretrained = range(1, len(history_resnet_pretrained['train_loss']) + 1)  # Epochenachse erzeugen
plt.figure(figsize=(10, 4))  # Plot-Canvas erzeugen
plt.subplot(1, 2, 1)  # Erstes Subplot fr Loss
plt.plot(epochs_range_pretrained, history_resnet_pretrained['train_loss'], label='Train-Loss')  # Train-Loss zeichnen
plt.plot(epochs_range_pretrained, history_resnet_pretrained['val_loss'], label='Val-Loss')  # Val-Loss zeichnen
plt.xlabel('Epoche')  # X-Achse beschriften
plt.ylabel('Loss')  # Y-Achse beschriften
plt.title('Pretrained ResNet50 Loss-Verlauf')  # Titel setzen
plt.legend()  # Legende anzeigen
plt.subplot(1, 2, 2)  # Zweites Subplot fr Accuracy
plt.plot(epochs_range_pretrained, history_resnet_pretrained['val_acc'], label='Val-Accuracy', color='tab:green')  # Val-Accuracy zeichnen
plt.xlabel('Epoche')  # X-Achse beschriften
plt.ylabel('Accuracy')  # Y-Achse beschriften
plt.title('Pretrained ResNet50 Validierungs-Accuracy')  # Titel setzen
plt.legend()  # Legende anzeigen
plt.tight_layout()  # Layout verdichten
plt.show()  # Plot anzeigen
```
